# Project: Climate Change Analysis in Tanzania

```python
# 1. Data Collection

# Import necessary libraries
# Data Handling
import pandas as pd
import numpy as np
import joblib
import os
import warnings
from datetime import datetime

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Time Series Decomposition
from statsmodels.tsa.seasonal import seasonal_decompose

# Machine Learning
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Suppress warnings
warnings.filterwarnings('ignore')
```

```python
# Load the climate dataset
df = pd.read_csv(r"D:\TANZANIA_KIC\capstone-project-nnko294\data\tanzania_climate_data.csv")
```

```python
# Standardize column names
df.columns = df.columns.str.strip().str.lower()
print("\nNormalized column names:", df.columns.tolist())

# Rename relevant columns for consistency
df.rename(columns={
    'average_temperature_c': 'temperature',
    'total_rainfall_mm': 'precipitation',
    'year': 'year',
    'month': 'month'
}, inplace=True)
```

```python
# Preview the dataset
df.head()
```

```python
# 2. Data Preprocessing

# Handle missing values (drop or impute as needed)
df = df.dropna()
```

```python
# Confirm 'year' and 'month' exist
assert 'year' in df.columns and 'month' in df.columns, "The dataset must contain 'year' and 'month' columns."
```

```python
# Create 'Season' based on 'month'
df['season'] = df['month'].map({12: 'Wet', 1: 'Wet', 2: 'Wet',
                                 3: 'Dry', 4: 'Dry', 5: 'Dry',
                                 6: 'Cool', 7: 'Cool', 8: 'Cool',
                                 9: 'Hot', 10: 'Hot', 11: 'Hot'})
```

```python
# One-hot encode 'season'
df = pd.get_dummies(df, columns=['season'], drop_first=True)
```

```python
# 3. Exploratory Data Analysis (EDA)

# Descriptive statistics
print(df.describe())
```

```python
# Plot average temperature by year
plt.figure(figsize=(10,5))
temp_by_year = df.groupby('year')['temperature'].mean()
temp_by_year.plot(label='Average Temperature')
plt.title('Average Temperature per Year')
plt.xlabel('Year')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()
```

```python
# Plot average precipitation by year
plt.figure(figsize=(10,5))
precip_by_year = df.groupby('year')['precipitation'].mean()
precip_by_year.plot(color='green', label='Average Precipitation')
plt.title('Average Precipitation per Year')
plt.xlabel('Year')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Seasonal decomposition (using synthetic monthly index)
monthly_temp = df.groupby(['year', 'month'])['temperature'].mean().reset_index()
monthly_temp['date'] = pd.to_datetime(monthly_temp[['year', 'month']].assign(day=1))
temp_series = monthly_temp.set_index('date')['temperature']
decomposition = seasonal_decompose(temp_series.dropna(), model='additive')
decomposition.plot()
plt.show()
```

```python
# 4. Machine Learning Model Development

# Define features and target
features = df.drop(columns=['temperature'])
target = df['temperature']
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Train a Random Forest model
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
# Evaluate model
print("Random Forest RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
print("Random Forest MAE:", mean_absolute_error(y_test, y_pred_rf))
```

```python
# Save the trained model for deployment
joblib.dump(rf, 'random_forest_model.pkl')
```
